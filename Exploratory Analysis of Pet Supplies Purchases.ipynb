## Analysis of Repeat Purchases

### Objective
The aim of this analysis is to compare products that have been repeatedly purchased with those that haven't. This comparison is based on the "repeat_purchase" column. The steps involved are as follows:

1. Calculate the number of products in each category for both repeated and non-repeated purchases.
2. Create a bar chart to visually represent the comparison results and identify trends.


**1. Calculating Product Counts in each category for both repeated and non-repeated purchases.**

```python
import pandas as pd

# Read the data from the CSV file
data = pd.read_csv(r'H:\Project\Exploratory Analysis of Pet Supplies Purchases\Data Source\updated_pet_supplies.csv')

# Group the data by 'category' and 'repeat_purchase', and calculate the count of 'product_id'
summary = data.groupby(['category', 'repeat_purchase'])['product_id'].count().reset_index()

# Rename the count column for clarity
summary.rename(columns={'product_id': 'product_count'}, inplace=True)

# Calculate the overall total count of 'product_id'
overall_total = data['product_id'].count()

# Create a row for the overall total in the summary DataFrame
overall_row = pd.DataFrame({'category': 'Overall', 'repeat_purchase': 'All', 'product_count': overall_total}, index=[0])

# Concatenate the overall row to the summary DataFrame
summary_with_total = pd.concat([summary, overall_row], ignore_index=True)

# Display the summary DataFrame
display(summary_with_total)
```

**2. Visualization: Comparison of Repeated and Non-repeated PurchasesComparison of Repeated and Non-repeated Pet Supplies Purchases by Category**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv(r'H:\Project\Exploratory Analysis of Pet Supplies Purchases\Data Source\updated_pet_supplies.csv')

# Group the data by 'category' and 'repeat_purchase', and calculate the count of 'product_id'
summary = data.groupby(['category', 'repeat_purchase'])['product_id'].count().reset_index()

# Pivot the data to create a pivot table
pivot_table = summary.pivot(index='category', columns='repeat_purchase', values='product_id').reset_index()

# Create a bar chart using Matplotlib
ax = pivot_table.plot(kind='bar',x='category', figsize=(10, 6))
plt.xlabel('Category')
plt.ylabel('Number of Product')
plt.title('Comparison of Repeated and Non-repeated Purchases by Category')
plt.xticks(rotation=45)
plt.legend(title='Repeat Purchase')

# Annotate the bars with the product counts
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

# Show the bar chart
plt.show()
```

## Insights from the Comparison of Repeated and Non-repeated Purchases by Category Bar Plot

**Key Observations:**

- **Repetition Trends:** Across all product categories, the highest product count is observed for repeated purchases, indicating consistent customer loyalty to repurchase.

- **Category Analysis:** The category "Equipment" has the highest product count for both repeated and non-repeated purchases, suggesting its popularity regardless of purchase type.

By recognizing the strong customer loyalty to repurchase and the popularity of "Equipment" products, the sales team can strategically focus on nurturing repeat customers and consider ways to leverage the appeal of "Equipment" items to drive sales across other categories.




## Sales Distribution Visualization

### Objective
The goal of this analysis is to understand the distribution of sales across different products. We achieved this through a histogram analysis, which provides insights into the spread and frequency of sales amounts. To understand the distribution of sales across different products, we generated a histogram analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv(r'H:\Project\Exploratory Analysis of Pet Supplies Purchases\Data Source\updated_pet_supplies.csv')

# Generate a histogram of the sales data
plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(data['sales'], bins=20, edgecolor='black')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Distribution of Sales across Products')
plt.grid(True)

# Annotate the bars with the product counts
for rect in patches:
    height = rect.get_height()
    plt.annotate(f'{int(height)}', xy=(rect.get_x() + rect.get_width() / 2, height),
                 xytext=(0, 3), textcoords="offset points",
                 ha='center', va='bottom', fontsize=10, color='black')

# Label x-axis with bin ranges
bin_width = bins[1] - bins[0]  # Calculate the width of each bin
bin_midpoints = [(bins[i] + bins[i+1]) / 2 for i in range(len(bins)-1)]
bin_labels = [f'{bins[i]:.2f} - {bins[i+1]:.2f}' for i in range(len(bins)-1)]
plt.xticks(bin_midpoints, bin_labels, rotation=45, ha='right')

plt.show()
```

## Insights from the Sales Distribution Histogram

The histogram analysis of sales distribution reveals:

- **Peak Sales Range:** Most of the sales are concentrated in the range of 976.10 to 1074.55, with a count of 325.
- **Sales Spread:** Sales are distributed across various ranges, indicating both high and low sales instances.
- **Outliers:** Some products stand out with exceptionally high or low sales, influencing the distribution.
- **Typical Sales:** The frequent sales range provides insight into the typical sales level for products.

By focusing on the peak sales range and identifying outlier products, the sales team can strategize effectively to boost sales within the most common range and address outliers for better overall performance.


## Sales and Repeat Purchase Relationship Analysis

### Objective
The goal of this analysis is to explore the relationship between sales and repeat purchase behavior using box plots. We aim to gain insights into how sales values differ between customers who make repeat purchases and those who do not. To achieve this, we utilize box plots to visualize the distribution and characteristics of sales data in relation to repeat purchase behavior.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data into a DataFrame
data = pd.read_csv(r'H:\Project\Exploratory Analysis of Pet Supplies Purchases\Data Source\updated_pet_supplies.csv')

# Create a figure with two subplots
plt.figure(figsize=(12, 6))

# Subplot 1: Box plot with outliers
plt.subplot(1, 2, 1)
sns.boxplot(x='repeat_purchase', y='sales', data=data)
plt.title('Sales and Repeat Purchase Relationship (With Outliers)')
plt.xlabel('Repeat Purchase')
plt.ylabel('Sales')

# Subplot 2: Box plot without outliers
plt.subplot(1, 2, 2)
sns.boxplot(x='repeat_purchase', y='sales', data=data, showfliers=False)
plt.title('Sales and Repeat Purchase Relationship (Without Outliers)')
plt.xlabel('Repeat Purchase')
plt.ylabel('Sales')

# Adjust layout
plt.tight_layout()
plt.show()

```

## Analyze the Shape of the Distribution and Identify Outliers

**Insights from the Sales and Repeat Purchase Relationship:**

**Key Observations:**

- **Sales Behavior Insights:**
Customers who don't make repeat purchases tend to have a wider range of sales amounts compared to those who do make repeat purchases. Additionally, the most extreme sales values are more common among customers who make repeat purchases.

- **Focus Areas for Action:**
Given the broader range of spending among non-repeat purchasers, there are opportunities to encourage them to make repeat purchases by tailoring offers or experiences to match their preferences. Addressing outliers in both customer groups can contribute to stabilizing sales trends. For repeat purchasers, particularly, dealing with these outliers is essential.

- **Outlier Impact:**
Although outliers can influence specific data points, the fundamental sales patterns for both groups remain consistent. Removing outliers doesn't substantially alter the typical sales levels. Thus, the focus should remain on the primary trends in sales behaviors.

By concentrating on these insights, the sales team can formulate strategies to engage non-repeat purchasers effectively, while also managing outliers to gain a more accurate understanding of customer dynamics and identify avenues for growth.
